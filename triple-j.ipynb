```python
import json
import networkx as nx
from pyvis.network import Network
```

```python
data = {}
with open('/Users/jonm/like-a-version.json') as f:
    data = json.load(f)
    print(data)
```

```python
artists = {}
for v in data.values():
    
    artist = v["artist_name"]
    if artist in artists:
        artists[artist]["raw_data"].append(v)
    else:
        artists[artist] = {"raw_data": [v]}

artists
```

```python
G = nx.Graph()

for artist, value in artists.items():
    G.add_node(artist, label=artist)
    for raw_data in value["raw_data"]:
        G.add_node(raw_data["subtitle"], label=raw_data["subtitle"])
        G.add_edge(artist, raw_data["subtitle"], label=raw_data['title'])

```

```python
nt = Network('2000px', '2000px', select_menu=True)
nt.from_nx(G)
# nt.toggle_physics(False)
nt.show('nx.html', notebook=False)
```

```python
path = nx.shortest_path(G, source="The Cardigans", target="Joji")
print(path)


print(sorted(nx.degree(G), key=lambda x: x[1], reverse=True))



```

```python
components = [G.subgraph(c).copy() for c in nx.connected_components(G)]
for c in sorted(components, key=lambda c: len(c.nodes()), reverse=True)[:5]:
    nt = Network('2000px', '2000px', select_menu=True)
    nt.from_nx(c)
    # nt.toggle_physics(False)
    nt.show(f'nx{len(c.nodes())}.html', notebook=False)
```
